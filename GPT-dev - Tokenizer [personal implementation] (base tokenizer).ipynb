```python
taylor = ""
with open('taylorswift.txt', 'r', encoding='utf-8') as f:
  taylor = f.read()

original = taylor
token_to_id = {char:idx for idx, char in enumerate(sorted(set(original)))}
id_to_token = {idx:char for idx, char in enumerate(sorted(set(original)))}

max_vocab_size = 2150
vocab_size = len(token_to_id)


def encode_text(text, vocabulary, train=False):
    # Split text into characters
    characters = list(text)
    encoded_output = []
    i = 0

    while i < len(characters):
        matched = False
        # Try to find the longest n-gram that fits
        for n in range(5, 0, -1):  # Assuming we have up to trigrams of characters
            if i + n <= len(characters):
                n_gram = ''.join(characters[i:i+n])
                if n_gram in vocabulary:
                    encoded_output.append(vocabulary[n_gram])
                    i += n  # Move past this n-gram
                    matched = True
                    break
    if train==True:
      encoded_output = [id_to_token[id] for id in encoded_output]

    return encoded_output

def decode_text(lista):
  tokens = [id_to_token[id] for id in encoded_output]
  return ''.join(tokens)

def generate_ngrams(text):
  encoded_text = encode_text(text, token_to_id, train=True) # since this is only used in training. might as well place this here
  concat = []
  n=2
  concat += [''.join(encoded_text[i:i+n]) for i in range(len(encoded_text) - n + 1) if (len(''.join(encoded_text[i:i+n])) <= 5)]
  return concat


from collections import Counter

while max_vocab_size >= vocab_size:
  total = generate_ngrams(original) # Output: ['ab', 'bc', 'cd', 'de', 'ef'] also concatenated with 3-grams, 4-grams, etc.
  filtered_ngrams = [x for x in total if x not in id_to_token.values()]
  if len(filtered_ngrams) == 0:
    break
  counter = Counter(filtered_ngrams)

  highest_count = max(counter.values())
  max_count = sum(1 for count in counter.values() if count == highest_count) # frequency of count for max-gram
  if highest_count == 1:
    break

  print(vocab_size)
  print("highest count: " + str(highest_count) + "\nmax_count: " + str(max_count))


  for i in range(highest_count // 10 + max_count + 10):
    if not (max_vocab_size >= vocab_size):
      break
    new_vocab = counter.most_common()[i][0]
    if new_vocab not in id_to_token.values():
      token_to_id[new_vocab] = vocab_size
      id_to_token[vocab_size] = new_vocab
      vocab_size += 1
    else:
      break

print(token_to_id)

print(encode_text(taylor, token_to_id, train=True))
```
